```python
import os
import yaml

# Define the directory and file path
new_dir = "/kaggle/working/yolov5/"
os.makedirs(new_dir, exist_ok=True)
yaml_path = os.path.join(new_dir, "dataset.yaml")

# Define the dataset structure
dataset = {
    'train': '/kaggle/input/datasetpair/dataset/images/train',   # Replace with your train images directory
    'val': '/kaggle/input/datasetpair/dataset/images/val',       # Replace with your validation images directory
    'nc': 2,                            # Number of classes
    'names': ['black', 'white']       # Class names
}

# Write the YAML file
with open(yaml_path, 'w') as file:
    yaml.dump(dataset, file, default_flow_style=False)

print(f"dataset.yaml written to: {yaml_path}")
```
